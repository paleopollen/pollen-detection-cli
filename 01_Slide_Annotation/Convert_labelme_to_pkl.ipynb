```python
# Set the project name and save directory for the project

project_name = 'Finetuning_allSlides'
trial = 'Trial_01' 

exp_dir = '/content/gdrive/MyDrive/ENSO_project/02_det/exp' # experiment directory
exp_dir = os.path.join(exp_dir, project_name)
save_dir = os.path.join(exp_dir, trial) 
if not os.path.exists(save_dir): 
    os.makedirs(save_dir)
```

```python
# do_not_include =['ind', 'spr', 'fun', 'fun?', 'fungal spore' 'what is this?']

center_radius_list = []

if os.path.exists(os.path.join(current_image_path, 'full_annot.plk')):
  with open(os.path.join(current_image_path,'full_annot.plk'), 'rb') as handle:
      tmpData_list = pickle.load(handle)
      print(tmpData_list)
      for i in range(len(tmpData_list)):
        shape = tmpData_list[i]
        label_name = shape["label"]
        # if label_name in do_not_include:
        #     continue
        points = shape['points']

        # # export crop masks
        xy = [tuple(point) for point in points]
        (cx, cy), (px, py) = xy
        d = math.sqrt((cx - px) ** 2 + (cy - py) ** 2)
        cx = round(cx)
        cy = round(cy)
        d = round(d)
        center_radius_list.append([label_name,'' , '', cx,cy,d])

print(center_radius_list)
```

```python
set_name='test'  #'test''train'
sampleList = dbinfo[set_name+'_det_list']

for i in range(len(sampleList)):
  current_example= sampleList[i] 
  current_image_path= os.path.join(path_to_image, current_example[0], current_example[1])

  annot = []

  if os.path.exists(os.path.join(current_image_path, 'full_annot.plk')):
    with open(os.path.join(current_image_path,'full_annot.plk'), 'rb') as handle:
        tmpData_list = pickle.load(handle)
        for j in range(len(tmpData_list)):
          shape = tmpData_list[j]
          label_name = shape["label"]
          # if label_name in do_not_include:
          #     continue
          points = shape['points']

          # # export crop masks
          xy = [tuple(point) for point in points]
          (cx, cy), (px, py) = xy
          d = math.sqrt((cx - px) ** 2 + (cy - py) ** 2)
          cx = round(cx)
          cy = round(cy)
          d = round(d)
          annot.append([label_name,'' , '', cx,cy,d])

  print(i, annot)
  save_dir = os.path.join(path_to_annot,current_example[0])
  if not os.path.exists(save_dir):
    os.makedirs(save_dir)

  with open(os.path.join(path_to_annot,current_example[0], current_example[1]+'.pkl' ), 'wb') as handle:
      pickle.dump(annot,handle)   
```
