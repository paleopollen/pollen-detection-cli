```python
ndpi_annot_dir = '/content/gdrive/MyDrive/ENSO_project/01_data/01_ndpi/annot_csv'
```

```python
os.listdir(ndpi_annot_dir)
```

```python
csv_list = []
for i in range(len(os.listdir(ndpi_annot_dir))):
  tmp_csv_path = os.path.join(ndpi_annot_dir, os.listdir(ndpi_annot_dir)[i])
  with open(tmp_csv_path , newline='', encoding= 'unicode_escape') as f:
    reader = csv.reader(f)
    data = list(reader)
    for j in range(len(data)):
      if j == 0:
        if i ==0:
          header = data[j]
        continue
      csv_list.append(data[j])

print(len(csv_list))
```

```python
print(csv_list)
```

```python
full_annot_path = os.path.join(ndpi_annot_dir, 'full_annot.csv')
with open(full_annot_path, "w", newline="", encoding= 'unicode_escape') as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerows(csv_list)
```

```python
with open(full_annot_path, newline='', encoding= 'unicode_escape') as f:
  reader = csv.reader(f)
  data = list(reader)

print(data)
```

```python
species_list = []
for i in range(len(data)):
  if i>0:
    code = data[i][3]
    species_list.append(code)

species_list = set(species_list)
```

```python
species_counts = {}
for i in range(len(csv_list)):
  key = csv_list[i][3]
  if key not in species_counts:
    species_counts[key] = 1
  else:
    species_counts[key] +=1

print(species_counts)
```

```python
sorted_species_counts = dict(sorted(species_counts.items()))
```
