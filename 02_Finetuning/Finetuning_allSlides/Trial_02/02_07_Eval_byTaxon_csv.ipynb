#count the number of gt grains for each class

```python
set_name = 'test'
sampleList = dbinfo[set_name+'_det_list']
```

```python
# idx = 10
iterCount = 0
codes_set = set()
codes_list = []
remove = ['ind','375', 'fun', 't58','t22','dia','ttr','ped','t47','p16','glo','p17','gel','spr','p2','neu','u4','u16','u17','his','u30','u45','peo','azo','bot']
no_annot = []

for idx in range(len(sampleList)):
    current_example = sampleList[idx]
    current_image_path= os.path.join(path_to_image, current_example[0], current_example[1])

    path_to_imdb = os.path.join(path_to_annot,current_example[0], current_example[1]+'.pkl')
    if not os.path.exists(path_to_imdb):
      no_annot.append(current_example)
      continue
    with open(os.path.join(path_to_annot,current_example[0], current_example[1]+'.pkl'), 'rb') as handle:
        annot = pickle.load(handle) 

    for i in range(len(annot)):
        code = annot[i][0]
        if code in remove:
            continue
        codes_set.add(code)
        codes_list.append(code)
    
    iterCount += 1

    if iterCount%25==0:
        print('{}/{}'.format(iterCount,len(sampleList)))

print(codes_set)
print(codes_list)
```

```python
gt_counts = {}
unique_codes = list(codes_set)
for j in range(len(unique_codes)):
  code = unique_codes[j]
  count = sum(x == code for x in codes_list)
  gt_counts.update({code:count})

print(gt_counts)
```

```python
count_compare = {}
gt_counts = gt_counts

for key in set(list(gt_counts.keys()) + list(det_counts.keys())):
    try:
        count_compare.setdefault(key,[]).append(gt_counts[key])        
    except KeyError:
        count_compare.setdefault(key,[]).append(0)
        pass

    try:
        count_compare.setdefault(key,[]).append(det_counts[key])          
    except KeyError:
        count_compare.setdefault(key,[]).append(0)
        pass

count_compare = dict(sorted(count_compare.items(), key=lambda item: item[0]))
print(count_compare)
```
